```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### B. Write Python code that performs (answers) the following operations (questions) 

##### 1. Read the data from the .csv file into a DataFrame and display the first five rows.

```python
# write your code here
data = pd.read_csv('avocado.csv')
data.head()
```

##### 2. Filter the data to exclude the following regions: TotalUS, West, and NorthEast. Use the filtered data to create a DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags columns grouped by type. Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado. Set the tile and axis labels of the bar plot and make sure all the labels are readable. 

```python
excluded_regions = ['TotalUS', 'West', 'NorthEast']
filtered_sum = data[~data['region'].isin(excluded_regions)].groupby('type').sum()

# Selecting only the required columns
result = filtered_sum[['Small Bags', 'Large Bags', 'XLarge Bags']]

# Plotting the data
result.plot.bar(title='Total Number of Small, Large, and XLarge Bags by Avocado Type')
plt.xlabel('Avocado Type')
plt.ylabel('Number of Bags')
plt.show()
```

##### 3. Analyze the avocado data from 2016 to 2018, to identify the top six regions where the sum of the average price of avocados is the highest. Create a visualization containing a grid of subplots, where each subplot is a box plot showing the distribution of average prices of organic and conventional avocados from 2016 to 2018, for the top six regions identified.

```python
# Filter the data to include only the years 2016 to 2018
data_filtered = data[(data['year'] >= 2016) & (data['year'] <= 2018)]

# Group the filtered data by region and calculate the sum of average prices
region_price = data_filtered.groupby('region')['AveragePrice'].sum()

# Select the top 6 regions with the highest sum of average prices
top_regions = region_price.nlargest(6).index

# Filter the data again to include only the top 6 regions
data_top = data_filtered[data_filtered['region'].isin(top_regions)]

# Separate the data into two DataFrames for organic and conventional avocados
organic_data = data_top[data_top['type'] == 'organic']
conventional_data = data_top[data_top['type'] == 'conventional']

# Create a 3x2 grid of subplots, making a total of 6 plots
fig, axs = plt.subplots(3, 2, figsize=(16, 9))
axs = axs.flatten()  # Flatten the array of axes for easier access

# Loop through each of the top 6 regions to create plots
for i, region in enumerate(top_regions):
    # Filter data for the current region for both organic and conventional
    organic_prices = organic_data[organic_data['region'] == region]['AveragePrice']
    conventional_prices = conventional_data[conventional_data['region'] == region]['AveragePrice']

    # Create a DataFrame to hold prices for both types for easier plotting
    prices_df = pd.DataFrame({
        'Organic': organic_prices,
        'Conventional': conventional_prices
    })

    # Create a box plot on the subplot at position i
    prices_df.boxplot(ax=axs[i])
    axs[i].set_title(region)  # Set title to the region's name
    axs[i].set_ylabel('Average Price')  # Label the y-axis

# Adjust layout to prevent overlap and display the plots
plt.tight_layout()
plt.show()
```
